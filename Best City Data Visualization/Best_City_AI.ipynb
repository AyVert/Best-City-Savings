# Best City to stay AI Generated Solution

Choose what city gives us the most savings over a 20 year period considering these factors. 
* The cities are "San Francisco", "New York", "Boston", "Oakland", "San Jose", "Los Angeles", "Washington", "Seattle", "San Diego", "Miami"
* The rent in each city is respectively [3500, 3000, 2590, 2500, 2450, 2260, 2260, 1890, 1790, 1800]
* Your standard yearly salary is 48,000
* You recieve a monthly bonus 3/10 of the city you are living in monthlys rent
* The bonus is only active for the first two years
* Groceries cost 1/100 of the city you are living in monthlys rent
* You are currently at Boston 
* To relocate to another city you must pay a $10,000 fee 
* All cities except Boston must pay the fee since you are already in Boston
* You are only allowed to move once
* you must live in each city for atleast a year
* List each city in order from greatest savings over 20 years to least
* Write a python code that will produce these results and plot them 


```python
import matplotlib.pyplot as plt

cities = ["San Francisco", "New York", "Boston", "Oakland", "San Jose", "Los Angeles", "Washington", "Seattle", "San Diego", "Miami"]
rent = [3500, 3000, 2590, 2500, 2450, 2260, 2260, 1890, 1790, 1800]
salary = 48000
bonus_multiplier = 3/10
groceries_multiplier = 1/100
bonus_active_years = 2
relocation_fee = 10000
years = 20

savings = []

# Calculate cumulative savings for each city
for i, city in enumerate(cities):
    cumulative_savings = 0
    current_city = "Boston"

    # Relocation fee for all cities except the current city (Boston)
    if city != current_city:
        cumulative_savings -= relocation_fee

    for year in range(1, years + 1):
        monthly_rent = rent[i]
        monthly_bonus = monthly_rent * bonus_multiplier if year <= bonus_active_years else 0
        monthly_groceries = monthly_rent * groceries_multiplier

        annual_savings = (salary - monthly_rent) * 12 + monthly_bonus * 12 - monthly_groceries * 12
        cumulative_savings += annual_savings

    savings.append(cumulative_savings)

# Sort cities based on cumulative savings
sorted_cities = [city for _, city in sorted(zip(savings, cities), reverse=True)]
sorted_savings = sorted(savings, reverse=True)

# Plot the savings
plt.bar(sorted_cities, sorted_savings)
plt.xlabel("City")
plt.ylabel("Cumulative Savings over 20 years ($)")
plt.title("Savings Comparison for Different Cities")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
